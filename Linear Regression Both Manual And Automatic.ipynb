# Linear Regression Type one (Manual)

```python
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([12,18,25,27,35])

n=len(x)

sum_x=np.sum(x)
sum_y=np.sum(y)
sum_x2=np.sum(x**2)
sum_xy=np.sum(x * y)
print("Sum x = ",sum_x)
print("Sum y = ",sum_y)
print("Sum x*x = ",sum_x2)
print("Sum xy = ",sum_xy)

m=(n* sum_xy- sum_x*sum_y) /(n* sum_x2-sum_x**2)
print("\nSlope (m) =",round(m,2))
x_mean=np.mean(x)
y_mean=np.mean(y)

c=y_mean - m* x_mean
print("Intercept (c) =", round(c,2))
print("\nRegession Equation: y ={:.2f}x + {:.2f}".format(m,c))
y_pred= m*x +c 
errors =y - y_pred
print("\nPredicted Values and Errors: ")
print("x y_actual y_pred error")
for i in range(n):
    print(x[i]," ",y[i],"    ",round(y_pred[i],2),"    " ,round(errors[i],2))
SSE=np.sum(errors**2)
MSE=SSE/n
print("\nSSE= ",round(SSE ,2))
print("MSE= ",round(SSE ,2))

MAE= np.mean(np.abs(errors))
print("MAE =",round(MAE,2))
SST=np.sum((y-y_mean)**2)
R2=1- (SSE/SST)
print("\nSST =", round(SST, 2)) 
print("R2 =",round(R2,4))
def predict(x_new):
    return m* x_new+c
print("\n----Prediction for new input----")
x_input=float(input("Enter Study Hours: "))
y_output=predict(x_input)
print(f"Predicted Marks ={y_output:.2f}")
```

# Linear Regression Type two (Automatic)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

### Data

```python
x=np.array([[1],[2],[3],[4],[5]])
y=np.array([12,18,25,27,35])
```

### Feature Scrolling

```python
scaler=StandardScaler()
X_scaled= scaler.fit_transform(x)
```

### Linear Regression

```python
model =LinearRegression()
model.fit(X_scaled, y )
```

### Predictions

```python
y_pred=model.predict(X_scaled)
```

### Evaluation

```python
print("Model Parameters: ")
print("Slope (n):",model.coef_[0])
print("Intercept (c):",model.intercept_)
print("\nEvaluation Metrics:")
print("MSF: ",mean_squared_error(y,y_pred))
print("MAE: ",mean_absolute_error(y,y_pred))
print("RMSE: ",np.sqrt(mean_squared_error(y,y_pred)))
print("R2: ",r2_score(y,y_pred))
```
