## First Machine Learning Code: 

### Concept used: Linear Regression

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
x = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(x,y)
prediction= model.predict([[10]])
print("Prediction for input 10 is: ",prediction)
```

### Taking Input

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
x = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(x,y)
uv = int(input("Enter the postion of x: "))
prediction= model.predict([[uv]])
print("Prediction for input",uv," is: ",prediction)
```

## Scatter plot and Data Points Example

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color= 'blue',label='Data Points')
plt.plot(x, model.predict(x),color='orange',label='Regression Line')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Linear Regression Line") 
plt.xlable("X")
plt.ylable("Y")
plt.tight_layout()
plt.show()
```
